```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.usedbooksfactory.com/buy-second-hand-old-books/category/best-sellers-books")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
categories = soup.find('section', attrs = {'id': 'aa-promo'})\
.find_all('a', class_ = 'related_category')
categories
```

```python
categories_list = []
```

```python
for i in categories:
  categories_list.append("https://www.usedbooksfactory.com" + i.get('href'))
```

```python
categories_list.pop(1)
```

```python
# Viewing the Category list
for i, category in enumerate(categories_list):
  print(i, category)
```

```python
for url in categories_list:
  new_res = requests.get(url)
  new_soup = BeautifulSoup(new_res.content, 'html.parser')

  # algorithm to find the Pagination
  try:
    last_page = int(new_soup.find('ul', class_ = 'pagination')\
    .find_all('li')[-2].text)
  except:
    last_page = 1

  # Looping through the url
  for page in range(1, last_page+1):
    # replace str(categories_list[:1]) with url
    res_ = requests.get(url + f"?page={page}")
    soup_ = BeautifulSoup(res_.content, "html.parser")

    # Entering into pages
    product_section = soup_.find('div', class_ = 'aa-product-catg-body').find_all('li')
    for product in product_section:
      book_title = product.find('img').get('alt')
      book_price = product.find('span', class_ = 'aa-product-price').text

      books.append([book_title, book_price])
```

```python
books
```

```python
books_data = books.copy()
```

```python
df = pd.DataFrame(books_data, columns = ['Book_title', 'Book_Price'])
```

```python
pd.set_option('display.max_rows', 10)
```

```python
df[['Book_title', 'Author']] = df['Book_title'].str.rsplit('-', 1, expand=True)
```

```python
df1 = df[['Book_title','Author','Book_Price']].copy()
```

```python
df1.rename(columns = {'Book_title': 'Book_Title'}, inplace = True)
```

```python
df1['Book_Title'] = df1['Book_Title'].str.replace('\d','', regex = True)\
.str.strip('#$')\
.str.strip()
```

```python
df1
```

```python
df1.rename(columns = {'Book_Price': 'Book_Price (in Rupees)'}, inplace = True)
```

```python
df1['Book_Price (in Rupees)'] = df1['Book_Price (in Rupees)'].str.replace('Rs', '')\
.str.strip()
```

```python
df1['Book_Price (in Rupees)'] = df1['Book_Price (in Rupees)'].astype('int')
```
